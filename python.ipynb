# Operators

```python
print("Arithmetic Operators")
# Arithmetic Operators
a, b = 10, 3

print(a + b)        # 13  → addition
print(a - b)        # 7   → subtraction
print(a * b)        # 30  → multiplication
print(a / b)        # 3.333... → division (always float)
print(a // b)       # 3   → floor division (integer result)
print(a % b)        # 1   → modulus (for remainder)
print(a ** b)       # 1000 → exponent (10^3)

print("Assignment Operators")
# Assignment Operators
x = 5
print(x)
x += 3         # x = x + 3 → 8
print(x)
x -= 2         # x = x - 2 → 6
print(x)
x *= 4         # x = x * 4 → 24
print(x)
x /= 3         # x = x / 3 → 8.0
print(x)
x //= 2        # x = x // 2 → 4.0    (still float, because previous was float)
print(x)
x %= 3         # x = x % 3 → 1.0
print(x)
x **= 2        # x = x ** 2 → 1.0
print(x)

print("Comparison Operators")
# Comparison Operators
print(5 == 5)            # True   → equal
print(5 != 3)            # True   → not equal
print(7 > 3)             # True   → greater than
print(2 < 10)            # True   → less than
print(5 >= 5)            # True   → greater than or equal
print(4 <= 2)            # False  → less than or equal

print("Logical Operators")
# Logical Operators
x = 10
print(x > 5 and x < 20)       # True  (both conditions True)
print(x > 15 or x < 20)       # True  (one condition True)
print(not(x > 5))             # False (negates True)

print("Bitwise Operators")
# Bitwise Operators
a = 5            # (binary: 0101)
b = 3            # (binary: 0011)

print(a & b)           # 1  (AND → 0001)
print(a | b)           # 7  (OR → 0111)
print(a ^ b)           # 6  (XOR → 0110)
print(~a)              # -6 (NOT → inverts bits)
print(a << 1)          # 10 (Left shift → 1010)
print(a >> 1)          # 2  (Right shift → 0010)

print("Membership Operators")
# Membership Operators
print("a" in "apple")         # True
print("z" not in "apple")     # True

nums = [1, 2, 3, 4]
print(3 in nums)              # True
print(10 not in nums)         # True

print("Identity Operators")
# Identity Operators
x = [1, 2, 3]
y = x
z = [1, 2, 3]

print(x is y)            # True (same object in memory)
print(x is z)            # False (same values, but different objects)
print(x == z)            # True (values are same)
print(x is not z)        # True
```

# Sets

Unordered, Unique Elements, Mutable

Creating Tuples

```python
# Using curly braces
my_set = {1, 2, 3}
print(my_set)

# Using set() function
my_set2 = set([1, 2, 2, 3])  # duplicates removed → {1, 2, 3}
print(my_set2)

# Empty set
s1 = set()           # correct
print(type(s1))
s2 = {}              # creates dictionary, not set
print(type(s2))
```

Set Methods & Operations

```python
# Adding & Removing

s = {1, 2}
s.add(3)             # {1, 2, 3} (Add single element)
print(s)

s.update([4, 5])     # {1, 2, 3, 4, 5} (Add multiple)
print(s)

s.remove(3)          # Removes 3, error if not found
print(s)

s.discard(10)        # Removes if present, no error
print(s)

s.pop()              # Removes random element
print(s)
print(s.pop())       # Removes & returns random element

s.clear()            # Removes all elements
print(s)
```

```python
# Mathematical Set Operations
A = {1, 2, 3}
B = {3, 4, 5}

print(A | B)   # Union → {1, 2, 3, 4, 5}
print(A & B)   # Intersection → {3}
print(A - B)   # Difference → {1, 2}
print(B - A)   # Difference → {4, 5}
print(A ^ B)   # Symmetric Difference → {1, 2, 4, 5}

# Equivalent methods:
print(A.union(B))                         # {1, 2, 3, 4, 5}
print(A.intersection(B))                 # {3} 
print(A.difference(B))                    # {1, 2} 
print(B.difference(A))                    # {4, 5} 
print(A.symmetric_difference(B))          # {1, 2, 4, 5} 
```

```python
# Set Relationship Checks
A = {3}
B = {3, 4, 5}

print(A.issubset(B))        # True if all elements of A in B
print(A.issuperset(B))      # True if all elements of B in A
print(A.isdisjoint(B))      # True if A & B have no common elements
```

```python
# Frozen Set
fs = frozenset([1, 2, 3])
# fs.add(4) - error (immutable)
print(fs)             # frozenset({1, 2, 3})
```

# Exception Handling

Try–Except Block

```python
try:
    x = 10 / 0
except ZeroDivisionError:
    print("You cannot divide by zero!")
```

Multiple Except Blocks

```python
try:
    num = int("abc")
except ValueError:
    print("Invalid conversion to int")
except ZeroDivisionError:
    print("Division by zero error")
```

Else Block

```python
try:
    x = 10 / 2
except ZeroDivisionError:
    print("Error")
else:
    print("Division successful:", x)
```

Finally Block

```python
try:
    f = open("test.txt", "w")
    f.write("Hello")
except Exception:
    print("Some error")
finally:
    f.close()
    print("File closed")
```

Example of all keywords:

```python
try:
    # Code that might cause an error
    num = int(input("Enter a number: "))
    print(10 / num)
except ZeroDivisionError:
    print("Cannot divide by zero!")
except ValueError:
    print("Invalid input! Please enter a number.")
else:
    print("No errors occurred!")  # Runs because no exception happened
finally:
    print("Execution complete.")
```

# Object-Oriented Programming (OOP)

```python
class Car:                               # Class
    def __init__(self, brand, color):    # Method
        self.brand = brand               # Attributes
        self.color = color

car1 = Car("Toyota", "Red")  # Object of 'Car' class
print(car1.brand)  # Toyota
```

Instance Variables vs Class Variables & Methods (Instance, Class, Static)

```python
class Student:
    school = "ABC School"  # Class attribute(Variable)

    def __init__(self, name):
        self.name = name  # Instance attribute (Variable)

    def instance_method(self):  # Instance Method
        print(f"Hello, I am {self.name}")

    @classmethod
    def class_method(cls):  # Class Method
        print(f"School Name: {cls.school}")

    @staticmethod
    def static_method():  # Static Method
        print("This is a static method")

# Using methods
s1 = Student("Alice")
s1.instance_method()      # Instance
Student.class_method()    # Class
Student.static_method()   # Static
```

Encapsulation

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner          # Public variable
        self._account_type = "Saving"  # Protected variable
        self.__balance = balance    # Private variable

    # Getter method (read balance)
    def get_balance(self):
        return self.__balance

    # Setter method (deposit money)
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
        else:
            print("Invalid deposit amount!")

    # Setter method (withdraw money)
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient funds or invalid amount!")

# Usage
acc = BankAccount("Unzila", 5000)
print(acc.owner)              # Public → Accessible
print(acc._account_type)      # Protected → Accessible but not recommended
# print(acc.__balance)        # Error: Private variable

print(acc.get_balance())      # Access via getter
acc.deposit(1000)             
print(acc.get_balance())      
acc.withdraw(2000)            
print(acc.get_balance())
```

Inheritance

```python
# Single

class Parent:
    def greet(self):
        print("Hello from Parent")

class Child(Parent):
    pass

obj = Child()
obj.greet()  # Output: Hello from Parent
```

```python
# Multiple

class A:
    def method_a(self):
        print("From A")

class B:
    def method_b(self):
        print("From B")

class C(A, B):
    pass

obj = C()
obj.method_a()
obj.method_b()
```

```python
# Multilevel

class Grandparent:
    def method_gp(self):
        print("From Grandparent")

class Parent(Grandparent):
    pass

class Child(Parent):
    pass

obj = Child()
obj.method_gp()
```

```python
# Hierarchical

class Parent:
    def greet(self):
        print("Hello from Parent")

class Child1(Parent):
    pass

class Child2(Parent):
    pass

obj1 = Child1()
obj2 = Child2()
obj1.greet()
obj2.greet()
```

```python
# Method Resolution Order (MRO)

print(C.__mro__)  # Shows the inheritance order
```

Method Overriding

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):  # Overriding the parent method
        print("Dog barks")

obj = Dog()
obj.speak()  # Output: Dog barks
```

Method Overloading (Not directly in Python)

```python
class Calculator:
    def add(self, a=0, b=0, c=0):  # Default arguments
        return a + b + c

calc = Calculator()
print(calc.add())           # Output: 0
print(calc.add(2, 3))       # Output: 5
print(calc.add(2, 3, 4))    # Output: 9
print(calc.add(5))          # Output: 5
```

Polymorphism

```python
class Dog:
    def sound(self):
        return "Woof"

class Cat:
    def sound(self):
        return "Meow"

# Using polymorphism
for animal in (Dog(), Cat()):
    print(animal.sound())  # Calls respective method for each object
```

Abstraction

```python
from abc import ABC, abstractmethod

class Shape(ABC):  # Abstract class
    @abstractmethod
    def area(self):
        pass  # Abstract method (no implementation here)

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height  # Implementation provided here

# shape = Shape()  # Cannot instantiate abstract class
rect = Rectangle(5, 10)
print(rect.area())  # Output: 50
```

super() Function

```python
class Parent:
    def __init__(self, name):
        self.name = name
        print("Parent init called")

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)  # Calls Parent's __init__
        self.age = age
        print("Child init called")

obj = Child("John", 20)
print(obj.name, obj.age)
```

# Advanced Functions & Concepts

Map, Filter, Reduce

```python
# Map

nums = [1, 2, 3]
squared = list(map(lambda x: x**2, nums))
print(squared)  # [1, 4, 9]
```

```python
# Filter

nums = [1, 2, 3, 4, 5]
even = list(filter(lambda x: x % 2 == 0, nums))
print(even)  # [2, 4]
```

```python
# Reduce

from functools import reduce

nums = [1, 2, 3, 4, 5]
sum_all = reduce(lambda x, y: x + y, nums)
print(sum_all)  # 15
```

Decorators

```python
# Simple Decorator

def decorator(func):
    def wrapper():
        print("Before function runs")
        func()          # original function
        print("After function runs")
    return wrapper

def say_hello():
    print("Hello!")

# Apply decorator
say_hello = decorator(say_hello)
say_hello()
```

```python
# Using @ Symbol (Pythonic Way)

def decorator(func):
    def wrapper():
        print("Before function runs")
        func()
        print("After function runs")
    return wrapper

@decorator            # @decorator is just a shortcut for say_hello = decorator(say_hello).
def say_hello():
    print("Hello!")

say_hello()
```

Generators & yield

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count   # Returns value and pauses the function
        count += 1

for num in count_up_to(5):
    print(num)
```

```python
class MyNumbers:
    def __init__(self):
        self.num = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.num <= 5:
            value = self.num
            self.num += 1
            return value
        else:
            raise StopIteration

nums = MyNumbers()
for n in nums:
    print(n)
```

Closures

```python
def outer_function(msg):
    def inner_function():
        print("Message:", msg)
    return inner_function

greet = outer_function("Hello")
greet()  # Still remembers 'msg' even though outer_function is done
```

Comprehensions with Conditions

```python
# List Comprehension
squares = [x**2 for x in range(10) if x % 2 == 0]
print(squares)
```

```python
# Set Comprehension
unique_lengths = {len(word) for word in ["apple", "banana", "kiwi", "apple"]}
print(unique_lengths)
```

```python
# Dict Comprehension
squares_dict = {x: x**2 for x in range(5) if x > 2}
print(squares_dict)
```

# Context Managers in Python

```python
class MyContext:
    def __enter__(self):
        print("Entering...")
        return "Resource Ready"
    
    def __exit__(self, exc_type, exc_value, traceback):
        print("Exiting... Cleanup done.")

# Using it
with MyContext() as resource:
    print(resource)
```

```python
# Contextlib Module (Shortcut)

from contextlib import contextmanager

@contextmanager
def my_context():
    print("Entering...")
    yield "Resource Ready"
    print("Exiting... Cleanup done.")

# Using it
with my_context() as resource:
    print(resource)
```

# Regular Expressions (re module)

```python
# Example with normal string:
import re

pattern = "\\d+"                                 # match one or more digits
print(re.findall(pattern, "My age is 21"))       # Output: ['21']
# Here we had to write double backslash \\d+, because \d alone is not valid in a normal string.
```

```python
# Example with raw string (better):
import re

pattern = r"\d+"   # raw string
print(re.findall(pattern, "My age is 21"))
# Output: ['21']
```

```python
import re

text = "My phone number is 9876543210"
pattern = r"\d{10}"  # 10-digit number

match = re.search(pattern, text)

if match:
    print("Found:", match.group())
```

```python
# Example without compile:
import re

text = "My phone is 12345 and your phone is 67890"

print(re.findall(r"\d+", text))
print(re.findall(r"\d+", text))
# Here, the pattern \d+ is compiled again and again.
```

```python
# Example with compile (better):
import re

pattern = re.compile(r"\d+")

text = "My phone is 12345 and your phone is 67890"

print(pattern.findall(text))
print(pattern.findall(text))
```

```python
# Examples
import re

# Find all digits
print(re.findall(r"\d+", "My number is 12345"))  
# ['12345']

# Check if a string starts with Hello
print(bool(re.match(r"Hello", "Hello World")))  
# True

# Replace spaces with hyphens
print(re.sub(r"\s+", "-", "Python is fun"))  
# Python-is-fun

# Extract email from text
text = "Contact me at test@example.com"
email = re.search(r"[a-zA-Z0-9._]+@[a-z]+\.[a-z]+", text)
print(email.group())  
# test@example.com
```
